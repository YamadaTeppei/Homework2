### 3.  アメリカ全土の統計情報のみを抜き出してみましょう. 

データのどこをみれば, 求めている情報が得られるのかを考え, 必要な関数を用いて表示する.

```python
totalUS = avocado[avocado["region"] == "TotalUS"]
totalUS
```

### 4. ニューヨーク地域における, アボカドの平均価格の年ごとの大まかな推移を, 折れ線グラフを用いて表してください.

CSVファイルには, 2015 - 2018 年のデータが含まれている.

```python
NewYork = avocado[avocado["region"]=="NewYork" ]
ave = NewYork.groupby("year").mean()["AveragePrice"]
print(ave)

x = ["2015", "2016", "2017",  "2018"]
y = [ave[2015], ave[2016], ave[2017], ave[2018]]

plt.plot(x, y, label="avocado_ave")
plt.legend()
plt.show()
```

### 5-B. #5-Aの比率の算出方法は正しいでしょうか? 正しくない場合, 新たな算出方法で比率を求めてみましょう.

```python
organic = totalUS[totalUS["type"] == "organic"]["Total Volume"].sum()
total = totalUS["Total Volume"].sum()
Percentage = organic / total * 100
print(Percentage)
```

### 6. #5を踏まえ, アメリカ全土を考えた際, 2016年と2018年を比べた場合, 売れたアボカドの種類の比率はどのように変化したと言えますか？

```python
organic_all = totalUS[totalUS["type"] == "organic"]
organic_2016 = organic_all[organic_all["year"] == 2016]["Total Volume"].sum()
organic_2018 = organic_all[organic_all["year"] == 2018]["Total Volume"].sum()

total_2016 = totalUS[totalUS["year"] == 2016]["Total Volume"].sum()
total_2018 = totalUS[totalUS["year"] == 2018]["Total Volume"].sum()

Percentage_2016 = organic_2016 / total_2016 * 100
Percentage_2018 = organic_2018 / total_2018 * 100
print("2016: ", Percentage_2016)
print("2018: ", Percentage_2018)
```

### 7. 2018年に, オーガニックアボカドを買った顧客数が最も多かったのはどの地域ですか?

必要に応じて行を削除することも重要な前処理.

```python
organic_2018all = avocado[avocado["year"] == 2018].loc[:,["Total Bags", "region"]]
groupe = organic_2018all.groupby("region").sum().drop("TotalUS")
max_value = groupe.max()
max_region = (groupe >= max_value)
max_region[max_region["Total Bags"]==True]
```

### 8. 2017年に最も多くのアボカドの売り上げをあげた地域はどこですか? グラフを用いて結論を導きだして下さい.

比較したのは, 売れたアボカドの**量**ではなく, **売り上げ**であることに注意する.

```python
avocado_2017 = avocado[avocado["year"] == 2017].loc[:, ["Total Volume", "AveragePrice", "region"]]
groupe = avocado_2017.groupby("region").sum().drop("TotalUS")
sales = groupe["Total Volume"] + groupe["AveragePrice"]
sales

```
